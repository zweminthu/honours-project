```python
import logging
import pandas as pd
import numpy as np
from numpy import random
import gensim
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import re
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('no_duplicate1.csv')
df = df[pd.notnull(df['Personality_Trait'])]
print(df.head(10))
print(df['Tweets'].apply(lambda x: len(x.split(' '))).sum())
```

```python
plt.figure(figsize=(10,4))
df.Personality_Trait.value_counts().plot(kind='bar');
```

```python
def print_plot(index):
    example = df[df.index == index][['Tweets', 'Personality_Trait']].values[0]
    if len(example) > 0:
        print(example[0])
        print('Personality_Trait', example[1])
print_plot(30)
```

```python
X = df.Tweets
y = df.Personality_Trait
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
cv = CountVectorizer(min_df=0., max_df=1.)
cv_matrix = cv.fit_transform(df['Tweets'])
cv_matrix = cv_matrix.toarray()
cv_matrix
```

```python
bv = CountVectorizer(ngram_range=(1,1))
bv_matrix = bv.fit_transform(df['Tweets'])

bv_matrix = bv_matrix.toarray()
vocab = bv.get_feature_names()
pd.DataFrame(bv_matrix, columns=vocab)
```

```python
tv = TfidfVectorizer(min_df=0., max_df=1., use_idf=True)
tv_matrix = tv.fit_transform(df['Tweets'])
tv_matrix = tv_matrix.toarray()

vocab = tv.get_feature_names()
pd.DataFrame(np.round(tv_matrix, 2), columns=vocab)
```
